```python
# 7. Do the necessary data pre-processing tasks on the previous result in order to get a clean dataframe with the following columns :
    # Asteroid ID
    # Asteroid name
    # The Minimal estimated diameter in Kilometre
    # Absolute_magnitude
    # Relative_velocity(km/s)

import requests
import pandas as pd

# Fetch and parse the NEO data
api_url = "https://api.nasa.gov/neo/rest/v1/feed?start_date=2025-05-12&end_date=2025-05-19&api_key=af5HNRxPmCNc0lUujvMMaWkES38fLCwTPNPhO1Vj"
responses1 = requests.get(api_url)

# Convert JSON response to Python dictionary
neo_data = responses1.json()

asteroids_list = []

# Extracting data
for date in neo_data['near_earth_objects']:
    for asteroid in neo_data['near_earth_objects'][date]:
        asteroids_data = {
            "Asteroid ID": asteroid.get("id"),
            "Asteroid name": asteroid.get("name"),
            "estimated_diameter_min": asteroid["estimated_diameter"]["kilometers"]["estimated_diameter_min"],
            "absolute_magnitude": asteroid.get("absolute_magnitude_h"),
            "relative_velocity(km/s)": float(asteroid["close_approach_data"][0]["relative_velocity"]["kilometers_per_second"])
        }
        asteroids_list.append(asteroids_data)

# Creating the DataFrame
asteroids_df = pd.DataFrame(asteroids_list)

# Displaying the first few rows of the DataFrame
print(asteroids_df.head())
```

```python
# 8. Try to export the new dataframe into a CSV file and share it with your colleagues

# Export DataFrame to CSV
asteroids_df.to_csv("asteroids_data.csv", index=False)

print("CSV file 'asteroids_data.csv' has been created successfully.")
```
